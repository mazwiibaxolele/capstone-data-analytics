### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
# Visualize the relationship between respondents' age and their participation frequency (SOPartFreq)
# Use the size of the bubbles to represent their job satisfaction (JobSat)

# Prepare the data
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65
}

df['Age_Numeric'] = df['Age'].map(age_mapping)
df['JobSat_Numeric'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Get participation frequency data (assuming SOPartFreq column exists)
if 'SOPartFreq' in df.columns:
    bubble_data = df[['Age_Numeric', 'SOPartFreq', 'JobSat_Numeric']].dropna()
    
    plt.figure(figsize=(12, 8))
    plt.scatter(bubble_data['Age_Numeric'], bubble_data['SOPartFreq'],
               s=bubble_data['JobSat_Numeric']*50, alpha=0.5,
               c=bubble_data['JobSat_Numeric'], cmap='viridis',
               edgecolors='black', linewidth=0.5)
    plt.colorbar(label='Job Satisfaction Score')
    plt.title('Age vs. Participation Frequency\n(Bubble Size = Job Satisfaction)')
    plt.xlabel('Age')
    plt.ylabel('Participation Frequency')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
# Visualize the relationship between yearly compensation (ConvertedCompYearly) and job satisfaction (JobSat)
# Use the size of the bubbles to represent respondents' age

df['CompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

bubble_comp = df[['CompYearly', 'JobSat_Numeric', 'Age_Numeric']].dropna()

plt.figure(figsize=(12, 8))
plt.scatter(bubble_comp['CompYearly'], bubble_comp['JobSat_Numeric'],
           s=bubble_comp['Age_Numeric']*10, alpha=0.5,
           c=bubble_comp['Age_Numeric'], cmap='coolwarm',
           edgecolors='black', linewidth=0.5)
plt.colorbar(label='Age')
plt.title('Compensation vs. Job Satisfaction\n(Bubble Size = Age)')
plt.xlabel('Yearly Compensation ($)')
plt.ylabel('Job Satisfaction Score')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nStatistics:")
print(f"Average Compensation: ${bubble_comp['CompYearly'].mean():,.2f}")
print(f"Average Job Satisfaction: {bubble_comp['JobSat_Numeric'].mean():.2f}")
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
# Visualize the popularity of programming languages respondents have worked with (LanguageHaveWorkedWith)
# across age groups. Use bubble size to represent the frequency of each language

if 'LanguageHaveWorkedWith' in df.columns:
    # Split and explode languages
    df_lang = df[['Age_Numeric', 'LanguageHaveWorkedWith']].copy()
    df_lang = df_lang.dropna()
    
    # Get top 10 languages
    all_languages = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()
    top_languages = all_languages.value_counts().head(10).index
    
    # Create data for bubble plot
    bubble_lang_data = []
    
    for lang in top_languages:
        for age_val in df['Age_Numeric'].dropna().unique():
            mask = (df['Age_Numeric'] == age_val) & (df['LanguageHaveWorkedWith'].str.contains(lang, na=False))
            count = mask.sum()
            if count > 0:
                bubble_lang_data.append({
                    'Language': lang,
                    'Age': age_val,
                    'Count': count
                })
    
    df_bubble_lang = pd.DataFrame(bubble_lang_data)
    
    plt.figure(figsize=(14, 8))
    
    # Create scatter plot with different colors for each language
    for lang in top_languages:
        lang_data = df_bubble_lang[df_bubble_lang['Language'] == lang]
        plt.scatter(lang_data['Age'], [lang] * len(lang_data),
                   s=lang_data['Count']*2, alpha=0.6,
                   label=lang, edgecolors='black', linewidth=0.5)
    
    plt.title('Programming Languages by Age\n(Bubble Size = Frequency)')
    plt.xlabel('Age')
    plt.ylabel('Programming Language')
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
# Explore the relationship between preferred databases (DatabaseWantToWorkWith) and job satisfaction
# Use bubble size to indicate the number of respondents for each database

if 'DatabaseWantToWorkWith' in df.columns:
    # Get top 10 databases
    all_databases = df['DatabaseWantToWorkWith'].dropna().str.split(';').explode()
    top_databases = all_databases.value_counts().head(10).index
    
    # Create data for bubble plot
    bubble_db_data = []
    
    for db in top_databases:
        mask = df['DatabaseWantToWorkWith'].str.contains(db, na=False)
        avg_jobsat = df[mask]['JobSat_Numeric'].mean()
        count = mask.sum()
        
        if not pd.isna(avg_jobsat):
            bubble_db_data.append({
                'Database': db,
                'AvgJobSat': avg_jobsat,
                'Count': count
            })
    
    df_bubble_db = pd.DataFrame(bubble_db_data)
    
    plt.figure(figsize=(12, 8))
    plt.scatter(range(len(df_bubble_db)), df_bubble_db['AvgJobSat'],
               s=df_bubble_db['Count']/2, alpha=0.6, c=df_bubble_db['AvgJobSat'],
               cmap='plasma', edgecolors='black', linewidth=1)
    plt.colorbar(label='Average Job Satisfaction')
    plt.xticks(range(len(df_bubble_db)), df_bubble_db['Database'], rotation=45, ha='right')
    plt.title('Preferred Databases vs. Job Satisfaction\n(Bubble Size = Number of Respondents)')
    plt.xlabel('Database')
    plt.ylabel('Average Job Satisfaction Score')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
# Visualize compensation (ConvertedCompYearly) across different developer roles (DevType)
# Use bubble size to represent job satisfaction

if 'DevType' in df.columns:
    # Get top 10 developer types
    top_devtypes = df['DevType'].value_counts().head(10).index
    
    bubble_dev_data = []
    
    for devtype in top_devtypes:
        devtype_data = df[df['DevType'] == devtype]
        avg_comp = devtype_data['CompYearly'].mean()
        avg_jobsat = devtype_data['JobSat_Numeric'].mean()
        count = len(devtype_data)
        
        if not pd.isna(avg_comp) and not pd.isna(avg_jobsat):
            bubble_dev_data.append({
                'DevType': devtype,
                'AvgComp': avg_comp,
                'AvgJobSat': avg_jobsat,
                'Count': count
            })
    
    df_bubble_dev = pd.DataFrame(bubble_dev_data)
    
    plt.figure(figsize=(14, 8))
    plt.scatter(range(len(df_bubble_dev)), df_bubble_dev['AvgComp'],
               s=df_bubble_dev['AvgJobSat']*100, alpha=0.6,
               c=df_bubble_dev['AvgJobSat'], cmap='viridis',
               edgecolors='black', linewidth=1)
    plt.colorbar(label='Job Satisfaction')
    plt.xticks(range(len(df_bubble_dev)), df_bubble_dev['DevType'], rotation=45, ha='right')
    plt.title('Compensation Across Developer Roles\n(Bubble Size = Job Satisfaction)')
    plt.xlabel('Developer Role')
    plt.ylabel('Average Compensation ($)')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
# Visualize the relationship between the collaboration tools used (NEWCollabToolsHaveWorkedWith) and age groups
# Use bubble size to represent the frequency of tool usage

if 'NEWCollabToolsHaveWorkedWith' in df.columns:
    # Get top 10 collaboration tools
    all_tools = df['NEWCollabToolsHaveWorkedWith'].dropna().str.split(';').explode()
    top_tools = all_tools.value_counts().head(10).index
    
    # Create data for bubble plot
    bubble_tool_data = []
    
    age_groups = sorted(df['Age_Numeric'].dropna().unique())
    
    for tool in top_tools:
        for age in age_groups:
            mask = (df['Age_Numeric'] == age) & (df['NEWCollabToolsHaveWorkedWith'].str.contains(tool, na=False))
            count = mask.sum()
            if count > 0:
                bubble_tool_data.append({
                    'Tool': tool,
                    'Age': age,
                    'Count': count
                })
    
    df_bubble_tool = pd.DataFrame(bubble_tool_data)
    
    plt.figure(figsize=(14, 10))
    
    # Create bubble plot
    for tool in top_tools:
        tool_data = df_bubble_tool[df_bubble_tool['Tool'] == tool]
        plt.scatter(tool_data['Age'], [tool] * len(tool_data),
                   s=tool_data['Count']*3, alpha=0.6,
                   edgecolors='black', linewidth=0.5)
    
    plt.title('Collaboration Tools by Age\n(Bubble Size = Frequency of Usage)')
    plt.xlabel('Age')
    plt.ylabel('Collaboration Tool')
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
# Explore the relationship between preferred web frameworks (WebFrameWantToWorkWith) and job satisfaction
# Use bubble size to represent the number of respondents

if 'WebFrameWantToWorkWith' in df.columns:
    # Get top 10 web frameworks
    all_frameworks = df['WebFrameWantToWorkWith'].dropna().str.split(';').explode()
    top_frameworks = all_frameworks.value_counts().head(10).index
    
    # Create data for bubble plot
    bubble_framework_data = []
    
    for framework in top_frameworks:
        mask = df['WebFrameWantToWorkWith'].str.contains(framework, na=False)
        avg_jobsat = df[mask]['JobSat_Numeric'].mean()
        count = mask.sum()
        
        if not pd.isna(avg_jobsat):
            bubble_framework_data.append({
                'Framework': framework,
                'AvgJobSat': avg_jobsat,
                'Count': count
            })
    
    df_bubble_framework = pd.DataFrame(bubble_framework_data)
    
    plt.figure(figsize=(14, 8))
    plt.scatter(range(len(df_bubble_framework)), df_bubble_framework['AvgJobSat'],
               s=df_bubble_framework['Count']/2, alpha=0.6,
               c=df_bubble_framework['AvgJobSat'], cmap='coolwarm',
               edgecolors='black', linewidth=1)
    plt.colorbar(label='Average Job Satisfaction')
    plt.xticks(range(len(df_bubble_framework)), df_bubble_framework['Framework'], 
              rotation=45, ha='right')
    plt.title('Preferred Web Frameworks vs. Job Satisfaction\n(Bubble Size = Number of Respondents)')
    plt.xlabel('Web Framework')
    plt.ylabel('Average Job Satisfaction Score')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
# Visualize the distribution of admired technologies (LanguageAdmired) across different countries (Country)
# Use bubble size to represent the frequency of admiration

if 'LanguageAdmired' in df.columns and 'Country' in df.columns:
    # Get top 5 countries and top 10 admired technologies
    top_countries = df['Country'].value_counts().head(5).index
    all_admired = df['LanguageAdmired'].dropna().str.split(';').explode()
    top_admired = all_admired.value_counts().head(10).index
    
    # Create data for bubble plot
    bubble_admired_data = []
    
    for country in top_countries:
        for tech in top_admired:
            mask = (df['Country'] == country) & (df['LanguageAdmired'].str.contains(tech, na=False))
            count = mask.sum()
            if count > 0:
                bubble_admired_data.append({
                    'Country': country,
                    'Technology': tech,
                    'Count': count
                })
    
    df_bubble_admired = pd.DataFrame(bubble_admired_data)
    
    plt.figure(figsize=(14, 10))
    
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    
    for idx, country in enumerate(top_countries):
        country_data = df_bubble_admired[df_bubble_admired['Country'] == country]
        tech_positions = [list(top_admired).index(tech) for tech in country_data['Technology']]
        
        plt.scatter(tech_positions, [country] * len(country_data),
                   s=country_data['Count']*5, alpha=0.6,
                   color=colors[idx], edgecolors='black', linewidth=0.5,
                   label=country)
    
    plt.xticks(range(len(top_admired)), top_admired, rotation=45, ha='right')
    plt.title('Admired Technologies Across Countries\n(Bubble Size = Frequency of Admiration)')
    plt.xlabel('Technology')
    plt.ylabel('Country')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```
